1. What is a parameter?
- Parameter: A parameter is a value that a machine learning model learns from the training data to make predictions.

- Example: In linear regression, the slope (weight) and intercept (bias) are parameters.

2. What is correlation?What does negative correlation mean?
- Correlation: Correlation measures the strength and direction of the relationship between two variables.

- Negative correlation: It means that when one variable increases, the other decreases.

- Example: As the price of a product increases, its demand usually decreases.

3. Define Machine Learning. What are the main components in Machine Learning?
- Machine Learning: Machine Learning is a branch of AI that enables computers to learn from data and make predictions or decisions without being explicitly programmed.

- Main components of Machine Learning:

  - Data – Used to train the model.
  - Features – Input variables used for prediction.
  - Model – Learns patterns from the data.
  - Algorithm – Method used to train the model.
  - Evaluation – Measures the model's performance.

4. How does loss value help in determining whether the model is good or not?
- Loss value measures the difference between the model's predicted output and the actual output.

  - Low loss → The model's predictions are close to the actual values, so the model is performing well.
  - High loss → The predictions are far from the actual values, so the model needs improvement.

- In short: Lower loss generally indicates a better model.

5. What are continuous and categorical variables?
- Continuous variables are numerical values that can be measured and broken down into decimals. They have an infinite number of possible values within a range.

  - Examples: Salary, Age, Temperature.

- Categorical variables represent distinct groups or labels (qualitative data). They have a fixed, finite number of categories.

  - Examples: Gender (Male/Female), Color (Red/Blue), Country.

6. How do we handle categorical variables in Machine Learning? What are the common techniques?
- Categorical variables are converted into numerical values because machine learning models cannot work directly with text.

 - Common techniques:

   - Label Encoding – Assigns a unique number to each category.
   - One-Hot Encoding – Creates a separate binary column for each category.
   - Ordinal Encoding – Used when categories have a natural order.

7. What do you mean by training and testing a dataset?
- Training is the process of feeding a specific portion of your data into an algorithm so it can discover and learn the underlying patterns and rules.
- Testing is the process of evaluating the trained model's performance on a separate, unseen portion of data to check how accurately it makes predictions.

9. What is a Test set?
- Test set: A test set is a portion of the dataset used to evaluate the performance of a trained machine learning model.

- It is not used during training and helps measure how well the model performs on new, unseen data.

  - Zero Correlation: A change in one variable does not affect the other.

15. What is causation? Explain difference between correlation and causation with an example.
- Causation: Causation means one variable directly causes a change in another variable.

- Difference:

   - Correlation means two variables are related.
   - Causation means one variable causes the other to change.

- Example:

   - Correlation: Ice cream sales and drowning cases both increase in summer, but ice cream does not cause drowning.
   - Causation: Pressing the brake pedal slows down the car. Here, pressing the brake causes the car to slow down.

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
- Optimizer: An optimizer is an algorithm that updates the model's parameters (weights) to minimize the loss and improve model performance.

- Common types of optimizers:

   - Gradient Descent (GD) – Updates weights using the entire dataset in each step.
   Example: Used for small datasets.
   - Stochastic Gradient Descent (SGD) – Updates weights using one training example at a time.
   Example: Faster for large datasets.
   - Mini-Batch Gradient Descent – Updates weights using a small batch of data.
   Example: Most commonly used in deep learning.
   - Adam (Adaptive Moment Estimation) – Combines the benefits of SGD and adaptive learning rates for faster and more stable training.
   Example: Widely used for neural networks.

- Feature scaling is the process of bringing all features to a similar scale or range.

- It helps because:

   - Prevents features with larger values from dominating.
   - Improves model accuracy.
   - Speeds up model training, especially for algorithms like KNN, SVM, and Gradient Descent.
